Question 2: What are the types of random variables?

**1. Discrete Random Variables**

  - Definition: These variables take on a finite or countably infinite set of distinct values (like 0, 1, 2, ...).
  - Key characteristic: There are "gaps" between possible values.
  - Distribution: Described by a Probability Mass Function (PMF) that assigns a probability to each value, with all probabilities summing to 1.
  - Examples:

     - Number of heads in coin tosses

      - Number of children in a family

       - Poisson-distributed counts

       - Binomial, Geometric, Bernoulli, Poisson distributions
**2. Continuous Random Variables**

  - Definition: These variables can take any value within a continuous interval (uncountably infinite). They have no gaps.
  - Probability behavior: The chance the variable equals an exact value is zero; instead, it's defined over intervals
  - Distribution: Described by a Probability Density Function (PDF); probabilities are integrals over a range
  - Examples:

     - Height of individuals

     - Time taken to complete a task

     - Rainfall amount, temperature, speed
**3. Mixed Random Variables**

 - Definition: These variables include both discrete and continuous components.
 - Illustration: For example, if you flip a coin and, if heads, spin a spinner to get a continuous value, but if tails, you assign a fixed number (say –1), the resulting variable is "mixed."

Question 3: Explain the difference between discrete and continuous distributions.

- **Discrete vs. Continuous Distributions**
- **Nature of Possible Outcomes**

  - Discrete Distributions

    - Random variables that take on countable values, such as whole numbers or distinct categories—like the number of heads in coin flips or the number of children in a family. These values are individual and separated.


  - Continuous Distributions

    - Random variables that take on an uncountable range of values, meaning they can be any real number within a range—like height, weight, or time. There are infinitely many possible values, even between any two given points.

**How Probabilities Are Assigned**

  - Discrete Distributions

    - Use a Probability Mass Function (PMF) that assigns a specific probability to each possible value. All these probabilities add up to 1.


  - Continuous Distributions

     - Use a Probability Density Function (PDF) instead. You cannot assign a positive probability to an exact value. Rather, probabilities are determined over intervals by integrating the density across that range.

- Cumulative Distribution Function (CDF)

   - Discrete

      - The CDF has jumps at each point where probability mass exists—because each individual outcome adds an atom of probability.



   - Continuous

       - The CDF is smooth and continuous, rising gradually and having no jumps.

Question 4: What is a binomial distribution, and how is it used in probability?

- The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent and identical trials, where each trial has only two outcomes—often termed success or failure. Each trial must have the same probability
𝑝
p of success. Examples include flipping coins, quality checks (defective or not), or pass/fail tests.

- When to Use the Binomial Distribution

   - The binomial distribution applies when the following conditions—known as Bernoulli trial assumptions—are met:

   - A fixed number 𝑛 of trials.

   - Trials are independent.

    - Each trial has only two outcomes: success or failure.

    - The probability of success p remains constant across trials.

- Practical Examples

   - Tossing a fair coin 10 times and finding the probability of exactly 6 heads.

    - Determining the chance of exactly 3 defective bulbs in a batch of 100, with each having a 2% defect probability.

     - Modeling how many patients respond positively when treatment success rate is known.

Question 7: What is the significance of confidence intervals in statistical analysis?

- Confidence intervals are vital in statistical analysis because they provide a range of plausible values for an unknown population parameter—such as a mean, proportion, or difference—rather than relying solely on a single point estimate. In effect, they quantify the precision of an estimate by incorporating sampling variability, with a wider interval indicating less certainty and a narrower one indicating more precision . A well-constructed confidence interval (typically 95%) has a concrete long-run probability interpretation: if an experiment were repeated numerous times, approximately 95% of those computed intervals would contain the true population value.By often excluding or including key values (like zero for differences), confidence intervals also implicitly inform about statistical significance, reflecting conclusions that mirror those produced by p-value–based hypothesis tests. Moreover, confidence intervals enhance interpretation by explicitly showing the magnitude and uncertainty of an effect, offering richer insight than a mere yes/no verdict of significance—making them especially valuable for assessing both practical and statistical relevance

- Precision and Uncertainty: Confidence intervals qualitatively and quantitatively express how reliable an estimate is—narrower intervals indicate higher precision, wider ones suggest caution

- Frequentist Interpretation: A 95% confidence level means that, in repeated sampling, roughly 95% of the calculated intervals would capture the true parameter

- Statistical Significance: If a confidence interval excludes the null value (e.g., zero for mean differences), this corresponds to a statistically significant result at the corresponding alpha level (e.g., 0.05)

Question 8: What is the concept of expected value in a probability distribution?
- The expected value of a random variable represents its long-term average outcome, weighted by the probability of each possible value. It serves as a central measure summarizing the distribution—just like an average summarizes a dataset.

- Center of distribution: Like a mean, the expectation reflects where a distribution “sits.” Whether it's graphically the center of the mass in a bell curve or the balancing point of discrete probabilities, it offers a useful summary measure

```python
'''
Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.
'''

import numpy as np
import matplotlib.pyplot as plt

# 1. Generate the data
mu = 50       # Desired mean
sigma = 5     # Desired standard deviation
size = 1000   # Number of random samples

data = np.random.normal(loc=mu, scale=sigma, size=size)  # Generates the normal samples

# 2. Compute sample statistics
sample_mean = data.mean()
sample_std = data.std(ddof=0)  # population standard deviation
# For sample standard deviation, you could use: data.std(ddof=1)

print(f"Sample Mean: {sample_mean:.3f}")
print(f"Sample Standard Deviation: {sample_std:.3f}")

# 3. Draw a histogram to visualize the distribution
plt.hist(data, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.title("Histogram of Normally Distributed Data (μ=50, σ=5)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
